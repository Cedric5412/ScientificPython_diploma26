### **Exercise 1:**
Write a function to solve a quadratic equation. Use the formula with the discriminant, only find the real solutions, if they don't exist print some message.

Hint: for the square root just use the power 0.5 to avoid having to call a special function which we haven't seen yet.

```python
def quadratic_solver(a, b, c):
    if a == 0:
        print('Not a quadratic equation! The root is x = ', -c/b)

    else:
        delta = b**2 - 4*a*c
        
        if delta < 0:
            print('No real solution!')
        elif delta == 0:
            x1 = -b/(2*a)
            print('Double solution x =', x1)
        else :
            x1 = (-b + delta**0.5) / (2*a)
            x2 = (-b - delta**0.5) / (2*a)
            print('Two solutions found:', x1, x2)
a = float(input('Enter a:'))
b = float(input('Enter b:'))
c = float(input('Enter c:'))
quadratic_solver(a, b, c)
```

### **Exercise 2:**
The Recamán sequence is a famous sequence invented by the Colombian mathematician, Bernardo Recamán Santos. It is defined by the following algorithm, starting at $a_0=0$:

$ a_n=a_{n-1}−n$ if $a_{n-1}-n > 0  $ and has not already appeared in the sequence; 

$ a_n=a_{n−1}+n$ otherwise

Write a function that returns the first N elements of this sequence.

You can use lists or sets to store the numbers. Time the computation

(to check that it works - you should get 0, 1, 3, 6, 2, 7, 13, 20, 12, 21, 11, 22, 10, 23, 9, 24, 8, 25, 43, 62, 42, 63, 41, 18, 42, 17, 43, 16, 44, 15, 45, 14, 46, 79, 113, 78, 114, 77, 39, 78, 38, 79, 37, 80, 36, 81, 35, 82)

```python
def recaman_sequence(n):
    seq = []
    seq.append(0)
    
    for i in range(1, n):
        a = seq[i-1] - i
        b = seq[i-1] + i
        
        if a > 0:
            if not a in seq:
                seq.append(a)
            else:
                 seq.append(b)
                
        else:
            seq.append(b)
            
    return seq
```

```python
%%timeit
recaman_sequence(20)
```

```python
def recaman_sequence_set(n):
    seq = []
    seq.append(0)
    
    for i in range(1, n):
        a = seq[i-1] - i
        b = seq[i-1] + i
        
        if a > 0:
            if not a in set(seq):
                seq.append(a)
            else:
                 seq.append(b)
                
        else:
            seq.append(b)
            
    return seq
```

```python
%%timeit
recaman_sequence(100)
```

```python
def recaman_sequence_set(n):
    seq = {0}
    val = 0
    
    for i in range(1, n):
        a = val - i
        b = val + i
        
        if a > 0:
            if not a in seq:
                seq.add(a)
                val = a
            else:
                seq.add(b)
                val = b
        else:
            seq.add(b)
            val = b
    return seq
```

```python
%%timeit
recaman_sequence_set(100)
```

### **Exercise 4:**
Given 2 lists, produce a list containing common items between the two.

```python
def list_intersection(a, b):
    intersect = []
    for i in a:
        if i in b:
            intersect.append(i)
    return list(set(intersect))
```

```python
a = [1,2,3,5,3,5,6]
b = [0,3,5,7,8,1]
list_intersection(a, b)
```

### **Exercise 5:**
Create a function that returns a list of all pairs of factors (as tuples) of a given number using list comprehension.

```python
def list_of_all_pairs(n):
    l = [(n//i, i) for i in range(1, n+1) if n % i == 0]
    return l
```

```python
list_of_all_pairs(36)
```

### **Exercise 7:**
Write a function to check if a string (containing spaces and capital letters, but no punctuation marks) is a palindrome (use help to see which functions are available for a string). 


```python
def is_palindrome(word):
    word = word.lower()
    word = list(word)
    for i in word:
        if i == ' ':
            word.remove(i)
    print(word)
    word2 = []
    for i in range(1, len(word)+1):
        word2.append(word[-i])
    if word == word2:
        return True
    else:
        return False
```

### **Exercise 8:**
Write a function to find a character with a maximum number of occurences in a string. Return a tuple with that character and the number. You can ignore ties. Only use string's inbuilt functions and the function `max` (use help to see how to use it).

```python
def max_occurrence(word):
    l = [word.count(i) for i in word]
    m = max(l)
    for i in word:
        if word.count(i) == m:
            return i, m
```

```python
word = 'Was it a CAR oR a Cat I sawh hh aaaaa    waxeaaaaaaaaaaaaaaaar'
max_occurrence(word)
```

### **Exercise 9:**
Write a function that return the first N prime numbers using the Sieve of Eratosphenes algorithm

https://en.wikipedia.org/wiki/Sieve_of_Eratosthenes

```python
def primes(n):
    A = [True]*(n+1)
    for i in range(2, int(n**0.5) + 1):
        B = [i**2 + k*i for k in range(n) if i**2 + k*i <= n]
        for j in B:
            A[j] = False
    return [i for i in range(len(A)) if A[i] == True and i>=2 and i <=n ]
```

```python
def n_first_primes(N):
    n0 = 2
    l = primes(n0)
    while len(l) < N:
        n0 = 2*n0
        l = primes(n0)
    return l[:N]
```

```python
n_first_primes(20), len(n_first_primes(20))
```
