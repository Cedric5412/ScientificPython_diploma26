Write a python program that solves the equation of motion for a frictionless physical
pendulum using the Euler and velocity-Verlet method: 

$$\frac{d^2x}{dt^2} = -\sin(x)$$

Solve with the initial conditions $t_0=0$, $x_0=1.0$, $v_0=0.0$ and $dt = 0.1$, $t_f=30$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f_x(t, x, y):
    return y

def f_y(t, x, y):
    return -np.sin(x)

def euler_method(t0, x0, y0, tf, dt):
    n = int(tf/dt)
    tlist = np.zeros(n+1)
    xlist = np.zeros(n+1)
    ylist = np.zeros(n+1)

    tlist[0] = t0
    xlist[0] = x0
    ylist[0] = y0

    for i in range(n):
        tlist[i+1] = tlist[i] + dt
        xlist[i+1] = xlist[i] + f_x(tlist[i+1], xlist[i], ylist[i])*dt
        ylist[i+1] = ylist[i] + f_y(tlist[i+1], xlist[i], ylist[i])*dt

    return tlist, xlist, ylist

def verlet_method(t0, x0, y0, tf, dt):
    n = int(tf/dt)
    tlist = np.zeros(n+1)
    xlist = np.zeros(n+1)
    ylist = np.zeros(n+1)

    tlist[0] = t0
    xlist[0] = x0
    ylist[0] = y0

    for i in range(n):
        tlist[i+1] = tlist[i] + dt

        # compte velocity first
        ylist[i+1] = ylist[i] + f_y(tlist[i+1], xlist[i], ylist[i])*dt/2

        # find the position
        xlist[i+1] = xlist[i] + ylist[i+1]*dt

        # update the velocity from new position
        ylist[i+1] = ylist[i+1] + f_y(tlist[i+1], xlist[i+1], ylist[i+1])*dt/2

    return tlist, xlist, ylist
```

```python
t0 = 0
x0 = 1.0 
y0 = 0.0
dt = 0.05
tf = 30
plt.figure(figsize=(10,5))
t, x, y = euler_method(t0, x0, y0, tf, dt)
tt, xx, yy = verlet_method(t0, x0, y0, tf, dt)

plt.scatter(t, x, label='Euler method')
plt.scatter(tt, xx, label='Verlet method')
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
plt.title('harmonic oscillator')
plt.legend()
```

**Non linear fitting**

From Numerical Methods I

```python
def b_lambda(l, b1, b2):
    return b1 / ( (l**5) * (np.exp(b2/l) - 1) )

def b_lambda_1(l, b1, b2):
    return 1 / ( (l**5) * (np.exp(b2/l) - 1) )

def b_lambda_2(l, b1, b2):
    return (-b1 * np.exp(b2/l)) / ( (l**6) * (np.exp(b2/l) - 1)**2 )
```

```python
def fit_nonlinear(x, y, b1, b2, alpha, e, max_it):
    n = np.size(x)
    r = np.zeros(n)
    J = np.zeros((n, 2))

    new_b1, new_b2 = b1, b2

    norm_delta_b = 10
    
    for i in range(max_it):
        for k in range(n):
            r[k] = y[k] - b_lambda(x[k], new_b1, new_b2)
            J[k, 0] = b_lambda_1(x[k], new_b1, new_b2)             
            J[k, 1] = b_lambda_2(x[k], new_b1, new_b2)   

        A = np.matmul(J.transpose(), J)
        
        A_inverse = np.linalg.inv(A)

        delta_b = np.matmul( np.matmul(A_inverse, J.transpose()), r)
        
        norm_delta_b = delta_b[0]**2 + delta_b[1]**2

        if norm_delta_b <= e:
            break

        new_b1 += alpha * delta_b[0]
        new_b2 += alpha * delta_b[1]

        var_e = 0

        for k in range(n):
            var_e = var_e + ((y[k] - b_lambda(x[k], new_b1, new_b2))**2)/(n-2)

        var_b = var_e * A_inverse
        se1 = np.sqrt(var_b[0,0])
        se2 = np.sqrt(var_b[1,1])

    return new_b1, new_b2, se1, se2
        
```

```python
b1 = 1.0                                           # Initial guess for b1
b2 = 1.0                                           #! Initial guess for b2
alpha = 0.1                                        # Damping factor
e = 1.0e-6                                          #! Convergence tolerance                                    
max_it = 1000
hc_kb = 14387.77

# Read data
def read_data(file_name):
    dataset = np.loadtxt(file_name)
    x = dataset[:, 0]
    y = dataset[:, 1]
    return x, y

x, y = read_data('data/sun_data.txt')

b1, b2, se1, se2 = fit_nonlinear(x, y, b1, b2, alpha, e, max_it)
print(b1)
print(b2)
print(se1)
print(se2)
print('Temperature of the sun:', hc_kb/b2)
```

```python
# Model
def model(x, beta1, beta2):
    return beta1 / ( (x**5) * (np.exp(beta2/x) - 1) )

plt.figure(figsize = (10, 6))
plt.scatter(x, y, marker = '.', color = 'darkslategrey', label = 'datapoint')
plt.plot(x, model(x, b1, b2), color ='crimson', label='fitted')
plt.xlabel('$\lambda$')
plt.ylabel('$B_{\lambda}$')
plt.legend()
plt.savefig('pictures/fitting.png')
```
